```python
import pandas as pd
import numpy as np
import bar_chart_race as bcr
import matplotlib.pyplot as plt
```

## Depuración DataSet

```python
df = pd.read_csv(r"C:\Users\Carla\Downloads\Dataset_extranjeros.csv", encoding='ISO-8859-1')
df
```

```python
df = df[["PAIS", "AÑO ESTIMACION" , "RRAA_REGULAR" ,"RRAA_IRREGULAR", "ESTIMACION"]]
df.head()
```

```python
df.sort_values("RRAA_IRREGULAR", ascending=False)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df2 = df.groupby(["AÑO ESTIMACION", "PAIS"]).sum()
df2
```

```python
df2 = df2.reset_index()
df2 = df2.set_index("AÑO ESTIMACION")
df2
```

```python
df2.columns
```

```python
df2.info()
```

```python
df_poblacion = df2[["ESTIMACION","PAIS"]]
df_poblacion.info()
```

```python
# Pivotar el DataFrame
df = df_poblacion.pivot_table(index='AÑO ESTIMACION', columns='PAIS', values='ESTIMACION')
df.head()
```

## Creación Bar Chart racing

```python
# Colores únicos
unique_colors = plt.cm.get_cmap('tab20', len(df.columns))
colors_list = [unique_colors(i) for i in range(len(df.columns))]

bcr.bar_chart_race(
    df=df,
    filename=r'C:\Users\Carla\Downloads\poblacion.gif', 
    orientation='h',
    sort='desc',
    n_bars=20,
    fixed_order=False,
    title='Estimación de Población extranjera en Chile',
    title_size=10,
    bar_label_size=8,
    tick_label_size=6,
    figsize=(6, 3.5),
    dpi=144,
    cmap=colors_list,
    period_length=1500, 
    steps_per_period=30  
)
```
